```python
import pandas as pd
import matplotlib.pyplot as plt

dados = []
for estudo in range(25):
    for experiencia in range(25):
        contrata = 0
        if estudo >=5:
            contrata = 1           
        else:
            if experiencia > 10 and estudo < 20:
                    contrata = 2     
            elif estudo < 20:
                    contrata = 3
       
        dados.append([estudo, experiencia, contrata])
                    
df = pd.DataFrame(dados, columns=['estudo','experiencia','contrata'])
scatter = plt.scatter(df.estudo,df.experiencia,c=df.contrata)

classes = ['Não contrata' if i==0 else f'Contrata {i}' for i in set(df.contrata) ]
plt.xlabel("Estudo")
plt.ylabel("Experiência")

plt.legend(handles=scatter.legend_elements()[0],loc='center left', labels=classes,bbox_to_anchor=(1, 0.5))
```
